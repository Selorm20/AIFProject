Import the libraries

```python
import gym
from minihack import LevelGenerator
import numpy as np
import matplotlib.pyplot as plt
from LibraryJacopo import remove_black_border
```

Create the level. By now, it is a 15x15 grid with some monsters and some lava cells, that are obstacles

```python
new_level = LevelGenerator(w=15,h=15)
new_level.set_start_pos((0,0))
# new_level.wallify()
new_level.add_monster(name='wolf',symbol='d', place=(1,2))
new_level.add_monster(name='coyote',symbol='d', place=(5,2))
new_level.add_monster(name='wolf',symbol='d', place=(3,8))
new_level.fill_terrain(type='fillrect',flag='L', x1=2,y1=2,x2=3,y2=3)
new_level.fill_terrain(type='fillrect',flag='L', x1=11,y1=2,x2=11,y2=10)
new_level.fill_terrain(type='fillrect',flag='L', x1=1,y1=13,x2=11,y2=13)
new_level.add_goal_pos((14,14))
print(new_level.get_des())
env = gym.make(
    "MiniHack-Skill-Custom-v0",
    des_file = new_level.get_des(),
    observation_keys=("chars", "pixel")
)
state = env.reset()
env.render()

plt.imshow(remove_black_border(state['pixel']))
plt.show()
```
